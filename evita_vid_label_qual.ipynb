```python
import pandas as pd
import numpy as np
import os
import glob
from pathlib import Path
```

```python
csv_files = glob.glob("/home/muhammad/Desktop/Datasets/dataset_evaluation_EVITA/newly_labeleld/*.csv")
csv_files = sorted(csv_files)
dest_dir_path = "/home/muhammad/Desktop/Datasets/dataset_evaluation_EVITA/disagreements_ann1_ann2"
```

```python
file_names_list=[]
total_snippets_list=[]
disagreed_snippets_list = []

snippets_all_csv = 0
disagreed_snippets_all_csv = 0


for csv_file_path in csv_files:
    this_file = os.path.basename(csv_file_path)
    file_names_list.append(this_file)
    anno_df = pd.read_csv(csv_file_path)
    
    if "source" in anno_df.columns:
        source_list=anno_df["source"].values
    elif "Source" in anno_df.columns:
        source_list=anno_df["Source"].values
        
    
    
    snippets_this_csv = len(source_list)
    snippets_disagreed_this_csv =0
    total_disagree_ann = 0
    
    for source in source_list:
        source_lower = str(source).lower()
        if "dis" in source_lower:
            snippets_disagreed_this_csv =snippets_disagreed_this_csv+1
        elif "ann" in source_lower:
            total_disagree_ann = total_disagree_ann+1 
            
    
    snippets_this_csv = snippets_this_csv - snippets_disagreed_this_csv + snippets_disagreed_this_csv/2
    
    snippets_all_csv = snippets_all_csv + snippets_this_csv
    disagreed_snippets_all_csv = disagreed_snippets_all_csv + snippets_disagreed_this_csv/2 + total_disagree_ann
    
    total_snippets_list.append(np.floor(snippets_this_csv))
    disagreed_snippets_list.append(np.floor(snippets_disagreed_this_csv/2 + total_disagree_ann ))
    
    
    print("file_name: " + this_file)
    print("total_snippets: "+ str(snippets_this_csv))
    print("Disagreed snippets: " + str(snippets_disagreed_this_csv/2 + total_disagree_ann ))
    print("---------------------------------------------------")
    
    
        
        
    
    
  

data_dict = {}
data_dict["file_name"] = file_names_list
data_dict["total_snippets"] = total_snippets_list
data_dict["disagreed_snippets"] = disagreed_snippets_list

df_export = pd.DataFrame.from_dict(data_dict)

dest_path = os.path.join(dest_dir_path, "ann1_ann2_disagreement.csv") 
df_export.to_csv(dest_path)
    
```

## From incorrect labeled Files

```python
def extract_data_from_End_col(label_df):
    filtered_info_collective = []
    col_name = label_df.columns[label_df.columns.str.contains(pat = 'End')]
    col_name = col_name[0]
    session_labels=label_df[col_name].to_list()
    for single_label in session_labels:
        single_label = str(single_label)
        comma_sep = single_label.replace(';', ',')
        comma_sep_list=comma_sep.split(',')
        #str_list = list(filter(None, comma_sep_list))
        str_list = ['NaN' if x=='' else x for x in comma_sep_list]
        #fileterd_info = str_list
        fileterd_info = [str_eleme.strip() for str_eleme in str_list]
        
        if len(fileterd_info) == 5:
        
            filtered_info_collective.append(fileterd_info)
        
        #print(fileterd_info)
        else:
            filtered_info_collective.append(['NaN', 'NaN', 'NaN', 'NaN', 'NaN'])
        
    return (col_name, filtered_info_collective)
```

```python
csv_files_labeled_incorrect = glob.glob("/home/muhammad/Desktop/Datasets/dataset_evaluation_EVITA/labeled_incorrect/*.csv")
```

```python
for sel_csv_incorr_lab in csv_files_labeled_incorrect:
    
    this_file = os.path.basename(sel_csv_incorr_lab)
    dest_dir = "/home/muhammad/Desktop/Datasets/dataset_evaluation_EVITA/newly_labeleld"
    dest_path = os.path.join(dest_dir, this_file)
    
    df_incorr = pd.read_csv(sel_csv_incorr_lab)
    col_name, filtered_info_collective = extract_data_from_End_col(df_incorr)
    col_name = ['source','label','start_n','end_n', 'remark']
    df_new_incorrect = pd.DataFrame(filtered_info_collective, columns= col_name)
    df_new_incorrect.to_csv(dest_path)
    
    
```
